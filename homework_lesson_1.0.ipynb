```python
import numpy as np 
import matplotlib.pyplot as plt

np.random.seed(2023)
a = np.random.randint(1,100,20)

print("array: ", a)

min_index = np.argmin(a)
print("The smallest number is at index: ", min_index)

plt.figure()
plt.plot(a)
plt.figure()
plt.plot(a, color='blue')

if min_index > 0:  
    plt.plot(range(min_index-1, min_index+1), a[min_index-1:min_index+1], color='red')
if min_index < len(a) - 1:  
    plt.plot(range(min_index, min_index+4), a[min_index:min_index+4], color='red')

plt.show()
```

```python
# Визначте максимальне значення для суми X + Y та добутку X * Y
max_sum = 100
max_product = 100

# Створіть словник для зберігання можливих добутків для кожної суми
sum_to_products = {}

# Переберіть всі можливі пари чисел (X, Y)
for X in range(2, max_sum):
    for Y in range(X+1, max_sum):
        # Обчисліть суму та добуток
        sum_xy = X + Y
        product_xy = X * Y

        # Якщо сума та добуток не перевищують максимальні значення,
        # додайте пару чисел (X, Y) до списку можливих пар для цього добутку
        if sum_xy <= max_sum and product_xy <= max_product:
            if product_xy not in sum_to_products:
                sum_to_products[product_xy] = []
            sum_to_products[product_xy].append((X, Y))

# Переберіть всі добутки та їх відповідні списки пар чисел
for product_xy, pairs in sum_to_products.items():
    # Якщо є лише одна можлива пара чисел для даного добутку,
    # це означає, що ми знайшли унікальне рішення
    if len(pairs) == 1:
        X, Y = pairs[0]
print((X, Y))
```
