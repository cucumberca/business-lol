```python
import numpy as np 
import matplotlib.pyplot as plt

np.random.seed(2023)
a = np.random.randint(1,100,20)

print("array: ", a)

min_index = np.argmin(a)
print("The smallest number is at index: ", min_index)

plt.figure()
plt.plot(a)
plt.figure()
plt.plot(a, color='blue')

if min_index > 0:  
    plt.plot(range(min_index-1, min_index+1), a[min_index-1:min_index+1], color='red')
if min_index < len(a) - 1:  
    plt.plot(range(min_index, min_index+4), a[min_index:min_index+4], color='red')

plt.show()
```

```python
#      X + Y   X * Y
max_sum = 100
max_product = 100

#         
sum_to_products = {}

#      (X, Y)
for X in range(2, max_sum):
    for Y in range(X+1, max_sum):
        #    
        sum_xy = X + Y
        product_xy = X * Y

        #        ,
        #    (X, Y)       
        if sum_xy <= max_sum and product_xy <= max_product:
            if product_xy not in sum_to_products:
                sum_to_products[product_xy] = []
            sum_to_products[product_xy].append((X, Y))

#         
for product_xy, pairs in sum_to_products.items():
    #          ,
    #  ,     
    if len(pairs) == 1:
        X, Y = pairs[0]
print((X, Y))
```
